# Самые популярные слова в каждой теме за 10 лет (10 на 1)

```python
from ted_talks.scraper import TEDScraper
import pandas as pd
import numpy as np
import random
import csv
import os
from IPython.display import clear_output
import time
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from matplotlib import pyplot as plt

topics = ['education', 'social change', 'culture', 'history', 'health', 'innovation', 'humanity', 
          'global issues', 'business', 'biology', 'communication', 'design', 'future', 'community', 
          'art', 'activism', 'creativity', 'climate change', 'environment', 'personal growth'] 

dataframe = pd.read_csv("dataset_to_work.csv")
dataframe = dataframe.dropna()

stopwords = set(STOPWORDS)
# снизу писать слова через запятую. Можно в двойных или в одинарных ковычках. 
# Чтобы все уместилось на экране, можно жать энтер после любой запятой и продолжать писать на следующей строке
# Те слова, что уже есть, можно стереть
stopwords.update(["want", "minutes", "two",
                  "think", "going", 'one'])

for topic in topics:
    d = dataframe.loc[dataframe["Topic"] == topic]
    text = ""
    for row in d.itertuples(index=False):
        text = text + " ".join(review for review in str((row._asdict())['Transcript']).split())

    wordcloud = WordCloud(width=1920, height=1080, stopwords=stopwords, background_color="white").generate(text)
    # Display the generated image:
    plt.figure(figsize=(20,10))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.show()
```
